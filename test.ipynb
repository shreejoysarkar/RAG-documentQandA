```python
import os
from app.config import get_settings
from app.core.embeddings import EmbeddingService

def check_api_key():
    try:
        settings = get_settings()
        print(f"Gemini API Key is set: {bool(settings.GEMINI_API_KEY)}")
        
        # Test embedding generation
        embedding_service = EmbeddingService()
        test_text = "This is a test query for checking API key validity."
        embedding = embedding_service.embed_query(test_text)
        print(f"API Key is valid. Embedding generated successfully. Length: {len(embedding)}")
        return True
    except Exception as e:
        print(f"API Key check failed: {str(e)}")
        return False

# Run the check
check_api_key()
```
